# NCES Data

```python
os.chdir(NCESPath)
```

```python
list(ncesFiles)
```

```python
ACT_Scores = pd.read_csv(ncesFiles[0])
Admissions = pd.read_csv(ncesFiles[1])
Degrees_byGender = pd.read_csv(ncesFiles[2])
Enrollment = pd.read_csv(ncesFiles[3])
Enrollment_byStudentLevel = pd.read_csv(ncesFiles[4])
FinancialAid = pd.read_csv(ncesFiles[5])
SAT_Scores = pd.read_csv(ncesFiles[6])
```

```python
nces = pd.merge(ACT_Scores, Admissions, on = "College Name")
nces = nces.drop(["UnitId_y"], axis = 1)
nces = nces.rename(columns = {'UnitId_x':'UnitId'})

nces1 = pd.merge(nces, Degrees_byGender, on = "College Name")
nces1 = nces1.drop(["UnitId_y"], axis = 1)
nces1 = nces1.rename(columns = {'UnitId_x':'UnitId'})

nces2 = pd.merge(nces1, Enrollment, on = "College Name")
nces2 = nces2.drop(["UnitId_y"], axis = 1)
nces2 = nces2.rename(columns = {'UnitId_x':'UnitId'})

nces3 = pd.merge(nces2, Enrollment_byStudentLevel, on = "College Name")
nces3 = nces3.drop(["UnitId_y"], axis = 1)
nces3 = nces3.rename(columns = {'UnitId_x':'UnitId'})

nces4 = pd.merge(nces3, FinancialAid, on = "College Name")
nces4 = nces4.drop(["UnitId_y"], axis = 1)
nces4 = nces4.rename(columns = {'UnitId_x':'UnitId'})

ncesFinal = pd.merge(nces4, SAT_Scores, on = "College Name")
ncesFinal = ncesFinal.drop(["UnitId_y"], axis = 1)
ncesFinal = ncesFinal.rename(columns = {'UnitId_x':'UnitId'})

list(ncesFinal)
```

# Merge NCCC and NCES

```python
ccData = pd.merge(ncccFinal, ncesFinal, on = "College Name")
list(ccData)
```
